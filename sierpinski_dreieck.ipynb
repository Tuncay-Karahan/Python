# Das Sierpinski Dreieck

```python
import matplotlib.pyplot as plt
```

### Definition der Sierpinski Dreieck's Funktion

```python
def sierpinski_triangle(ax, points, depth):
    if depth == 0:
        triangle = plt.Polygon(points, edgecolor='black')
        ax.add_patch(triangle)
    else:
        midpoints = [
            [(points[0][0] + points[1][0]) / 2, (points[0][1] + points[1][1]) / 2],
            [(points[1][0] + points[2][0]) / 2, (points[1][1] + points[2][1]) / 2],
            [(points[2][0] + points[0][0]) / 2, (points[2][1] + points[0][1]) / 2]
        ]
        
        sierpinski_triangle(ax, [points[0], midpoints[0], midpoints[2]], depth - 1)
        sierpinski_triangle(ax, [points[1], midpoints[0], midpoints[1]], depth - 1)
        sierpinski_triangle(ax, [points[2], midpoints[1], midpoints[2]], depth - 1)
```

## Bestimmen der Iterationsvorschrift und Plotteinstellungen

```python
# Set the figure size
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')
ax.axis('off')

# Define the initial triangle vertices
points = [[0, 0], [1, 0], [0.5, 0.866]]

# Recursion depth
depth = 10

sierpinski_triangle(ax, points, depth)

plt.show()
```
