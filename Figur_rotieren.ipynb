```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_min, x_max, y_min, y_max = -4, 4.5, -5, 7.5
```

```python
fig, ax = plt.subplots(figsize=((x_max-x_min)/1.5,(y_max-y_min)/1.5))
```

```python
ax.set_aspect('equal')
display(fig)
```

```python
ax.axis([x_min,x_max,y_min,y_max])
display(fig)
```

```python
x_ticks = np.arange(x_min,x_max+0,1)
print('x_ticks = ',x_ticks)
ax.set_xticks(x_ticks)
y_ticks = np.arange(y_min,y_max+0,1)
print('y_ticks = ',y_ticks)
ax.set_yticks(y_ticks)
display(fig)
```

```python
ax.grid(True)
display(fig)
```

```python
from matplotlib.ticker import AutoMinorLocator
```

```python
ax.set_xticks(np.arange(x_min,x_max),minor=True)
ax.set_yticks(np.arange(y_min,y_max),minor=True)
ax.minorticks_on()
ax.grid(which='minor', alpha=0.1)
ax.grid(which='major', alpha=0.5)
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
display(fig)
```

```python
x1 = np.array([-3,0])
x2 = np.array([0,3])
x3 = np.array([-2,2])
y1 = np.array([-3,6])
y2 = np.array([6,-3])
y3 = np.array([0,0])
#ax.plot(x,y,'-k')
ax.plot(x1,y1,'-k')
ax.plot(x2,y2,'-k')
ax.plot(x3,y3,'-k')
display(fig)
```

```python
alpha = 20*np.pi/180
R = np.array([[np.cos(alpha),-np.sin(alpha)],[np.sin(alpha),np.cos(alpha)]])
platzhalter = np.array([np.nan])
x = np.concatenate((x1,platzhalter,x2,platzhalter,x3))
y = np.concatenate((y1,platzhalter,y2,platzhalter,y3))
xy = np.array([x,y])

Rxy = R@xy

rx = Rxy[0,:]
ry = Rxy[1,:]

```

```python
ax.plot(rx,ry,'r')
display(fig)
```
