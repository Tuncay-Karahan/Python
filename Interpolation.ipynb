```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_min, x_max, y_min, y_max = 0, 6.5, -1, 2.5
```

```python
fig, ax = plt.subplots(figsize=((x_max-x_min),(y_max-y_min)))
```

```python
ax.set_aspect('equal')
display(fig)
```

```python
ax.axis([x_min,x_max,y_min,y_max])
display(fig)
```

```python
x_ticks = np.arange(x_min,x_max+0,1)
print('x_ticks = ',x_ticks)
ax.set_xticks(x_ticks)
y_ticks = np.arange(y_min,y_max+0,1)
print('y_ticks = ',y_ticks)
ax.set_yticks(y_ticks)
display(fig)
```

```python
ax.grid(True)
display(fig)
```

```python
from matplotlib.ticker import AutoMinorLocator
```

```python
ax.set_xticks(np.arange(x_min,x_max),minor=True)
ax.set_yticks(np.arange(y_min,y_max),minor=True)
ax.minorticks_on()
ax.grid(which='minor', alpha=0.1)
ax.grid(which='major', alpha=0.5)
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
display(fig)
```

```python
ax.scatter(1,0,marker = "*",color = 'r')
ax.scatter(2,0,marker = "*",color = 'r')
ax.scatter(3,1,marker = "*",color = 'r')
ax.scatter(4,0,marker = "*",color = 'r')
ax.scatter(5,0,marker = "*",color = 'r')
display(fig)
```

```python
x = np.array([1,2,3,4,5])
A = np.transpose(np.array([x**0,x**1,x**2,x**3,x**4]))
print(A)
b = np.array([0,0,1,0,0])
a = np.linalg.solve(A,b)
x = np.linspace(0,6.5)
y = a[0]+a[1]*x+a[2]*x**2 +a[3]*x**3 +a[4]*x**4
ax.plot(x,y,'-b')
display(fig)
```
